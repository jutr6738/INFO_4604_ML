# Example 1: Simple Linear Regression

```python
#Import matplotlib

import matplotlib.pyplot as plt


#Declare some data...
x = [5,7,8,7,2,17,2,9,4,11,12,9,6]
y = [99,86,87,88,111,86,103,87,94,78,77,85,86]

#Show scatterplot of data
plt.scatter(x, y)
plt.show()
```

```python
#Fit a linear model on data

import matplotlib.pyplot as plt
from scipy import stats

x = [5,7,8,7,2,17,2,9,4,11,12,9,6]
y = [99,86,87,88,111,86,103,87,94,78,77,85,86]

#obtain regression outputs with stats.linregress...
slope, intercept, r, p, std_err = stats.linregress(x, y)

def myfunc(x):
  return slope * x + intercept

mymodel = list(map(myfunc, x))

#display scatterplot of original data
plt.scatter(x, y)

#plot "best fit" line (using parameters obtained from stats.linregress) on scatterplot of original data
plt.plot(x, mymodel)

#display line and scatterplot
plt.show()
```

```python
#Execute a method that returns some important key values of Linear Regression:

#stats.linregress returns the following:

#1. slope of the regressor
#2. intercept
#3. correlation coefficient between the outcome variable (regressand) and explanatory variable (regressor)
#4. p-value of the null hypothesis test that there is no statistically significant relationship between outcome and regressor
#. standard error of the regression


#The standard error of the regression, also known as the standard error of the estimate, 
#represents the average distance that the observed values fall from the regression line. 
#Conveniently, it tells you how wrong the regression model is on average using the units of the response/outcome variable


#In an analogy to standard deviation, taking the square root of MSE yields the root-mean-square error or root-mean-square deviation (RMSE or RMSD), 
#which has the same units as the quantity being estimated; 
#for an unbiased estimator, the RMSE is the square root of the variance, known as the standard error



slope, intercept, r, p, std_err = stats.linregress(x, y)
```

```python
#Its important to note that Statsmodels does not add a constant (i.e. intercept) by default. 
#Lets see it first without a constant in our regression model:
```

```python
## Case 1: Without a constant

import statsmodels.api as sm

X = df["RM"]
y = target["MEDV"]

# Note the difference in argument order
model = sm.OLS(y, X).fit()


# Print out the statistics
model.summary()
```

```python
#Make predictions using the model

predictions = model.predict(X) # make the predictions by the model
```

```python
#A regression without a constant implies that the regression line should run through the origin, 
#i.e., the point where both the response variable and predictor variable equal zero

#You should typically not exclude a constant term in yor SLR. Including a constant term helps make 
# the coefficient estimates unbiased, so you want that benefit. Unbiased estimates (and therefore models)
# can be generalized unlike the case when you have biased coefficient estimates
```

```python
#Case 2: Simple Linear Regression with a constant

#If we do want to add a constant to our model  we have to set it by using the command X = sm.add_constant(X) 
#where X is the name of your data frame containing your input (independent) variables.
```

```python
import statsmodels.api as sm # import statsmodels 

X = df["RM"] ## X usually means our input variables (or independent variables)
y = target["MEDV"] ## Y usually means our output/dependent variable
X = sm.add_constant(X) ## let's add an intercept (beta_0) to our model

# Note the difference in argument order
model = sm.OLS(y, X).fit() ## sm.OLS(output, input)


# Print out the statistics
model.summary()
```

```python
predictions = model.predict(X)
```

```python
predictions
```

# Example A - simple Multiple Linear Regression (MLR) example 

```python
import pandas as pd

#create DataFrame
df = pd.DataFrame({'hours': [1, 2, 2, 4, 2, 1, 5, 4, 2, 4, 4, 3, 6],
                   'prep_exams': [1, 3, 3, 5, 2, 2, 1, 1, 0, 3, 4, 3, 2],
                   'score': [76, 78, 85, 88, 72, 69, 94, 94, 88, 92, 90, 75, 96]})

#view DataFrame
print(df)
```

```python
#Obtain ordinary least squares (OLS) estimate of slope

lm.coef_
```

```python
#Obtain and print ordinary least squares (OLS) estimate of intercept

lm.intercept_
```

```python
#Obtain and print coefficient of determination (R-squared)

lm.score(X,y)
```

# Example: y = b0 + bx + bx^2 + e

```python
Well use 3x4  7x3 + 2x2 + 11 to improve your polynomial related vocabulary with some essential definitions:

degree of a polynomial: the highest power (largest exponent) in your polynomial; in our example its 4 because of x4, meaning that were dealing with a 4th degree polynomial,
coefficient: each number (3, 7, 2, 11) in our polynomial is a coefficient; these are the parameters that are unknown and our polynomial regression model will try to estimate when trained on our dataset,
leading term: the term with the highest power (in our case its 3x4); this is the most important part of a polynomial, because it determines the polynomials graph behavior,
leading coefficient: the coefficient of the leading term (its 3 in our polynomial),
constant term: the y intercept, it never changes: no matter what the value of x is, the constant term remains the same (in this case it is 11).
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.arange(0, 30)
y = [3, 4, 5, 7, 10, 8, 9, 10, 10, 23, 27, 44, 50, 63, 67, 60, 62, 70, 75, 88, 81, 87, 95, 100, 108, 135, 151, 160, 169, 179]
plt.figure(figsize=(10,6))
plt.scatter(x, y)
plt.show()
```

```python
#Changing an array from 1D to other dimensions
x = np.arange(0, 30).reshape(-1,1)
x
```

```python
#Step 1: Specifying the degree of the polynomial

#Import PolynomialFeatures
from sklearn.preprocessing import PolynomialFeatures
```

```python
#Then save an instance of PolynomialFeatures with the following settings
#PolynomialFeatures helps set the degree of the model - highest power of the regressor

poly = PolynomialFeatures(degree=2, include_bias=False)
```

```python
#degree sets the degree of our polynomial function. degree=2 means that we want to work with a 2nd degree polynomial

#include_bias=False should be set to False, because well use PolynomialFeatures together with LinearRegression() later on


#Long story short: LinearRegression() will take care of this setting by default, so theres no need to set include_bias to True.
#If it wasnt taken care of, then include_bias=True would mean that we 
#deliberately want the y intercept (0) to be equal to 0  but we dont want that (since we want unbiased estimates).
#Unbiased estimates help us generalize our models to new data

#Except you want a regression through the origin (0)
```

```python
#the following code confirms you have created an instance of PolynomialFeatures
poly
```

```python
#Step 2: Creating the new features/predictors

poly_features = poly.fit_transform(x.reshape(-1, 1))
```

```python
#Show that you have created two features - X and X^2

poly_features
```

```python
#reshape(-1,1) transforms our numpy array x from a 1D array to a 2D array  this is required, 
#otherwise we would get an error
```

```python
#Theres only one method  fit_transform()  but in fact its an amalgam of two separate methods: 
#fit() and transform(). fit_transform() is a shortcut for using both at the same time, because theyre often used together
```

```python
#Step 3: Creating the polynomial regression model. Import LinearRegression from sklearn

from sklearn.linear_model import LinearRegression
```

```python
#Save an instance of LinearRegression to a variable/object

poly_reg_model_1 = LinearRegression()
```

```python
#Fit model to data

poly_reg_model_1.fit(poly_features, y)
```

```python
#Fitting means that we train our model by letting it know what the feature (poly_features) and the response (y) values are. 
#When fitting/training our model, we basically instruct it to solve for the coefficients (marked with bold) in our polynomial function
```

```python
#Now that our model is properly trained, we can put it to work by instructing it to predict the responses (y_predicted)
#based on poly_features, and the coefficients it had estimated:

y_predicted_1 = poly_reg_model_1.predict(poly_features)
```

```python
#View predicted values

y_predicted_1
```

```python
#Lets do some dataviz to see what our model looks like:

plt.figure(figsize=(10, 6))
plt.title("Your first polynomial regression  congrats! :)", size=16)
plt.scatter(x, y)
plt.plot(x, y_predicted_1, c="red")
plt.show()
```

```python
#Since there is no train/test split, we do not obtain a RMSE
```

```python
#Obtain and print Polynomial model coefficient estimates 
poly_reg_model_1.coef_
```

```python
#Obtain and print Polynomial model intercept

poly_reg_model_1.intercept_
```

```python
#Obtain and print coefficient of determination (R-squared)

poly_reg_model_1.score(poly_features,y)
```

# Coding a polynomial regression model with multiple features

# Example: y = b0 + b1x_1 + b2x_2 + b3x_1^2 + b4x_2^2 + b5x_1x_2 + e

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Step 1: Simulate dataset

np.random.seed(1)#Set random seed to drive reproducibility
x_1 = np.absolute(np.random.randn(100, 1) * 10)
x_2 = np.absolute(np.random.randn(100, 1) * 30)
y = 2*x_1**2 + 3*x_1 + 2 + np.random.randn(100, 1)*20
```

```python
#Above we create some random data with some added noise: 
#x_1 contains 100 values for our first feature, 
#x_2 holds 100 values for our second feature. 
#the responses (100) are saved into the response variable y, as always.
```

```python
#Visualize a 2D chart of the features

fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10, 4))
axes[0].scatter(x_1, y)
axes[1].scatter(x_2, y)
axes[0].set_title("x_1 plotted")
axes[1].set_title("x_2 plotted")
plt.show()
```

```python
#Step 2: Storing the variables in a dataframe

df = pd.DataFrame({"x_1":x_1.reshape(100,), "x_2":x_2.reshape(100,), "y":y.reshape(100,)}, index=range(0,100))
```

```python
#display dataframe

df
```

```python
#Again, reshape(100,) is needed (100, because we have 100 rows), otherwise wed receive an error
```

```python
#Step 3: Define the training and the test data

#import train and test split
from sklearn.model_selection import train_test_split

#X, y = df[["x_1", "x_2"]], df["y"]: From df, we save the x_1 and x_2 columns to X, the y column to y.
#At this point we have the features and the responses stored in separate variables.

X, y = df[["x_1", "x_2"]], df["y"]


#poly = PolynomialFeatures(degree=2, include_bias=False) and poly_features = poly.fit_transform(X): 
#we create the new polynomial features.

poly = PolynomialFeatures(degree=2, include_bias=False)

#Within the train_test_split method we define all of our features (poly_features) and all of our responses (y). 
#Then, with test_size we set what percentage of our all features (poly_features) and responses (y) 
#wed like to allocate for testing our models prediction capability.


poly_features = poly.fit_transform(X)


#random_state needs to receive an integer value: that way anytime you rerun train_test_split, youll receive the same results

X_train, X_test, y_train, y_test = train_test_split(poly_features, y, test_size=0.3, random_state=42)
```

```python
#X_train, X_test, y_train, y_test: train_test_split splits our features (poly_features) and responses (y) into train and test groups  here we just save these groups into variables. 
#The order of the variables is very important, so dont shuffle them.
```

```python
#Step 4: Create a polynomial regression model

#Now lets create and fit our model, but this time, well train our model only on the training data:

poly_reg_model = LinearRegression()
poly_reg_model.fit(X_train, y_train)
```

```python
#The reason for us training our model only on the training data is that later we want to see 
#how well our model will predict responses based on feature values it hasnt seen before
```

```python
#Heres how we can test how our model performs on previously unseen data:


poly_reg_y_predicted = poly_reg_model.predict(X_test)

from sklearn.metrics import mean_squared_error

#Obtain model RMSE
poly_reg_rmse = np.sqrt(mean_squared_error(y_test, poly_reg_y_predicted))

#display model RMSE

poly_reg_rmse
```

```python
#display Polynomial model RMSE

poly_reg_rmse
```

```python
#Obtain and print Polynomial model coefficient estimates 
poly_reg_model.coef_
```

```python
#Obtain and print Polynomial model intercept

poly_reg_model.intercept_
```

```python
#We take the square root of mean_squared_error to get the RMSE (root mean square error) 
#which is a commonly used metric to evaluate a machine learning models performance. 
#RMSE shows how far the values your model predicts (poly_reg_y_predicted) are from the true values (y_test), on average. 
#Roughly speaking: the smaller the RMSE, the better the model.
```

```python
#Poly_features contains new features created out of x_1 and x_2, 
#so our polynomial regression model (based on a 2nd degree polynomial with two features) looks like this:

y = 0 + 1x1 + 2x2 + 3x1^2 + 4x2^2 + 5x1x2

#This is because poly.fit_transform(X) added three new features to the original two (x1 (x_1) and x2 (x_2)): x12, x22 and x1x2.
```
